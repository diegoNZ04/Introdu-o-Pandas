Importando Pandas

```python
import pandas as pd
```

Transformando Dicionário em DataFrame

```python
#DataFrame = pd.DataFrame()
vendas = {
    'data': ['15/02/2021', '16/02/2021'] ,
    'valor': [500, 300],
    'produto:': ['feijão', 'arroz'],
    'qnte': [50, 70]
}

tabela_vendas = pd.DataFrame(vendas)
display(tabela_vendas)
```

Método .loc
- Pegar 1 linha
- Pegar linhas de acordo com alguma condição
- Pegar linhas e colunas específicas
- Pegar 1 valor específico

```python
#Pegar 1 linha
display(vendas_df.loc[1])
```

```python
#Pegar um escopo de linhas
display(vendas_df.loc[1:5])
```

```python
#Pegar linhas de acordo com uma condição
#vendasnorteshopping = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping']
#display(vendasnorteshopping)
```

```python
#Pegar várias linhas e colunas
# .loc[linhas, colunas] -> parâmetros

vendasnorteshoppingformat = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping', ['ID Loja', 'Produto', 'Quantidade']]
display(vendasnorteshoppingformat)
```

```python
#Pegar um valor específico
print(vendas_df.loc[1, 'Produto'])
```

Adicionar 1 Coluna

```python
#a partir de uma coluna que existe
vendas_df['Comissão'] = vendas_df['Valor Final'] * 0.05
display(vendas_df)
```

```python
#criar coluna com valor padrão
vendas_df.loc[:, 'Imposto'] = 0
display(vendas_df)
```

Adicionar 1 Linha
- linhas de um complemento de base de dados

```python
vendas_dez_df = pd.read_excel('Vendas - Dez.xlsx')
#display(vendas_dez_df)

vendas_df = pd.concat([vendas_df, vendas_dez_df])
display(vendas_df)
```

Excluir Linhas e Colunas

```python
vendas_df = vendas_df.drop(0, axis=0)
display(vendas_df)
```

Calcular Indicadores
- Groupby
- Value Conts

```python
#value conts
transacoes_loja = vendas_df['ID Loja'].value_counts()
display(transacoes_loja)
```

```python
#group by
faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum()
display(faturamento_produto)
```

Mesclar 2 DataFrames (Procuror Informações de um DataFrame em outro)

```python
gerentes_df = pd.read_excel('Gerentes.xlsx')
display(gerentes_df)
```

```python
vendas_df = vendas_df.merge(gerentes_df)
display(vendas_df)
```
